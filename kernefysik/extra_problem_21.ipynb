```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(-np.pi, 4*np.pi, 100)

def y_1(xs):
    return np.cos(xs/2) + np.sin(xs/2)

def y_2(xs):
    return np.cos(xs/2) - np.sin(xs/2)

plt.plot(xs, y_2(xs), label="y2")
plt.plot(xs, y_1(xs), label="y1")
plt.legend()
plt.show()
```
