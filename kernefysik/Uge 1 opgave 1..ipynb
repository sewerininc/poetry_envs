```python
def semi_emperic(A: int, Z: int, b_v = 15.2, b_s = 15.8, b_symm = 44.7, b_c = 0.686, b_p = 5.0) -> float:
    """
    all constants in MeV
    Returns binding energy
    """
    N = A - Z
    output = b_v * A - b_s * A**(2/3) - b_symm * (N - Z)**2 / (2 * A) - b_c * Z**2 / A**(1/3) + b_p * ((-1)**N + (-1)**Z) * 1 / A**(1/2)
    return output / A 
    
    
```

```python
import pandas as pd
import re

def load_mass_list(file_path):
    # Read data from the file, skipping the first 30 rows
    with open(file_path, 'r') as file:
        # Skip the first 30 rows
        for _ in range(34):
            next(file)
        # Read the remaining data
        data = file.read()

    # Split the input data into lines
    lines = data.split('\n')

    # Define column names
    columns = [1,2,3,4,5,6,7,'N', 'Z', 'A', 'El', 'Orig', 'Mass Excess', 'Binding Energy/A', 'Beta-decay Energy', 'Atomic Mass']

    # Initialize lists to store data
    rows = []

    # Iterate through each line
    for line in lines:
        # Skip empty lines
        if not line.strip():
            continue

        # Split the line into tokens
        tokens = re.split(r'\s+', line.strip())

        # Check if the line contains numerical values (indicating a new row)
        if any(token.replace('.', '', 1).isdigit() for token in tokens):
            # Start a new row
            current_row = tokens
            rows.append(current_row)
        else:
            # Concatenate the current line to the existing row
            current_row += tokens

    # Create a DataFrame using the collected data
    df = pd.DataFrame(rows, columns=columns)

    return df
```

```python
# Example usage:
file_path = "massround.mas20.txt"
df = load_mass_list(file_path)
```
