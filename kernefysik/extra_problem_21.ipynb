```python
E1 = 0
E2 = np.linspace(2, -2, 100)
V_int = 0.3


sqrt_val = np.sqrt((E2 - E1)**2 + 4*V_int**2)

E_tot_p = (E1 + E2) + sqrt_val
E_tot_m = (E1 + E2) - sqrt_val

def c(E, E_tot, V):
    return 1 / (1 + (E_tot / 2 - E)**2 / V**2)

#plt.plot(E2, c(E1, E_tot_p, V_int), label="c1: Eplus")
#plt.plot(E2, c(E2, E_tot_p, V_int), label="c2: Eplus")
plt.plot(E2, c(E1, E_tot_m, V_int), "--" , label="$|C_{1-}|^2$ (Closed Shell)")
plt.plot(E2, c(E2, E_tot_m, V_int), "--", label="$|C_{2-}|^2$ (Open Shell)")
plt.plot(E2, c(E1, E_tot_m, V_int) + c(E2, E_tot_m, V_int), label="$|C_{2-}|^2 + |C_{1-}|^2$", color="black")
plt.legend()
#plt.yscale("log")
plt.xlabel("E' [MeV]")
plt.ylabel("Probability")
plt.ylim(0,1.05)
plt.show()
```

```python
plt.plot(E2, E_tot_p, label="E_plus")
plt.plot(E2, E_tot_m, label="E_minus")
plt.plot([-2, 2],[0,0], "--", color="black", label="Unperturbed energies")
plt.plot([-2, 2],[-4,4], "--", color="black")
plt.annotate("E_plus", [1.7,3.1])
plt.annotate("E_minus", [1.5,0.2])
plt.xlabel("2E' [MeV]")
plt.ylabel("Eigenvalues [MeV]")
plt.legend() # Skriv upertuberede løsninger ind 
```

For de upertuberede energier (V = 0) er E_plus = 2E' og E_minus = 0 

Når (upertuberede) E_plus = E_min er delta E mellem de to pertuberede energi niveauer 2V

```python
def Emarks(E_mark):
    E_plus = 1.058
    Vr = 0.3 
    output = (E_mark + np.sqrt(E_mark**2 + 4*Vr**2))/2 - E_plus
    return output
```

```python
xs = np.linspace(0,1,20)
plt.plot(xs, Emarks(xs))
```

```python
fsolve(Emarks, 0.3)[0]
```
