```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import sph_harm

# Define the range of theta
theta = np.linspace(0, np.pi, 100)

# Compute the spherical harmonics for the states |2, 0>, |2, 1>, and |2, 2>
Y_20 = np.abs(sph_harm(20, 20, 0, theta))**2
Y_21 = np.abs(sph_harm(40, 40, 0, theta))**2
Y_22 = np.abs(sph_harm(5, 5, 0, theta))**2

# Plotting
plt.plot(theta, Y_20, label='$|2, 0\\rangle$')
plt.plot(theta, Y_21, label='$|2, 1\\rangle$')
plt.plot(theta, Y_22, label='$|2, 2\\rangle$')

plt.xlabel('$\\theta$')
plt.ylabel('Probability Density')
plt.title('Probability Density for States $|2, 0\\rangle$, $|2, 1\\rangle$, and $|2, 2\\rangle$')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
from scipy.integrate import quad
from scipy.special import sph_harm

# Define the probability density function
def probability_density(theta, phi, m, j):
    return np.abs(sph_harm(m, j, phi, theta))**2

# Define the integrand for <cos^2 theta>
def integrand(theta, phi, m, j):
    return probability_density(theta, phi, m, j) * np.cos(theta)**2 * np.sin(theta) *2*np.pi

# Integrate the integrand over the interval [0, pi]
cos2_expectation_20, _ = quad(integrand, 0, np.pi, args=(0,0, 2))
cos2_expectation_21, _ = quad(integrand, 0, np.pi, args=(0,1, 2))
cos2_expectation_22, _ = quad(integrand, 0, np.pi, args=(0,2, 2))

print("<cos^2 theta> for |2, 0>: ", cos2_expectation_20)
print("<cos^2 theta> for |2, 1>: ", cos2_expectation_21)
print("<cos^2 theta> for |2, 2>: ", cos2_expectation_22)
```
